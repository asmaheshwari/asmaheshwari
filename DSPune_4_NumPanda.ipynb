```python
import numpy as np
x= np.array([3,5,7,2,8,10,11,65,72,81,99,100,150])
k=x.size
y=x.copy()

n=int(input("Enter Window Value"))
K=k-n+1
for i in range(K):
    y[i]=0
    for j in range(i,i+n,1):
        y[i] += x[j]
    print("Y["+str(i)+"] = ", round(y[i]/n,2))
```

```python
# Find the sum of the values in s for every Wednesday
# Generic program to enter day of the week which will return sum of values of that day of week.
import numpy as np
import pandas as pd
x=input('Enter Day of the Week')
days=pd.date_range(start='2015-01-01', end='2016-01-01', freq='D')
val=np.random.randint(1,100,366)
df=pd.DataFrame({'Date' : days, 'Value' : val})
df['weekday'] = df['Date'].dt.weekday_name
wd=df.groupby('weekday')
wd.groups
wd.describe()
df_wd=wd.sum()
print(df_wd)
print()
print("Sum on {} is {}".format(x,wd.get_group(x).sum()))
```

```python
# 4. Average For each calendar month
import numpy as np
days=pd.date_range(start='2015-01-01', end='2016-01-01', freq='D')
val=np.random.randint(1,100,366)
df=pd.DataFrame({'Date' : days, 'Value' : val})
df['Month_Number'] = df['Date'].dt.month
df['Month'] = df['Date'].dt.month_name()
df_month=df.groupby(['Month_Number','Month'])
round(df_month.mean(),2)
```

```python
# 5. # For each group of four consecutive calendar months in s, find the date on which the highest value occurred.
import pandas as pd
import numpy as np

days=pd.date_range(start='2015-01-01', end='2015-12-31', freq='D')
val=np.random.randint(1,100,365)
df=pd.DataFrame({'Date' : days, 'Value' : val})
df['Month_Number'] = df['Date'].dt.month
df_max = df[df['Value']==df['Value'].max()]
df_max
```
